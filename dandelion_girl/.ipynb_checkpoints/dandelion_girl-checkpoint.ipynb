```python
import re 
import string
import PyPDF2
```

```python
# creating a pdf file object 
pdfFileObj = open('young-robert-f-the-dandelion-girl.pdf', 'rb') 
  
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 

pages = []

for page in range(pdfReader.numPages):
    pages.append(pdfReader.getPage(page).extractText())
    
# closing the pdf file object 
pdfFileObj.close()
pages
```

```python
f=open('pdf.txt','w')
for page in range(len(pages)):
    f.write(pages[page])
f.close()
```

```python
with open('pdf.txt') as f:
    clean = f.read().splitlines()
```

```python
clean
```

```python
shear = [i.replace('-','') for i in clean ]
shear = [re.sub(r'[^\w\s]', "", s) for s in shear] # get rid of punctuation
shear = [re.sub(r'\\w*\\d\\w*', "", s) for s in shear] # get rid of numerics

shears = [x for x in shear if x != ' ']
shearss = [x for x in shears if x != '']
```

```python
shear
```

```python
shears
```

```python
shearss
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd
import numpy as np
%pylab
%matplotlib inline
```

```python
vect = TfidfVectorizer(ngram_range=(1,1),stop_words='english')
```

```python
dtm = vect.fit_transform(shearss)
```

```python
dtm
```

```python
pd.DataFrame(dtm.toarray(),columns=vect.get_feature_names())
```

```python
lda = LatentDirichletAllocation(n_components=20)
```

```python
lda_dtf = lda.fit_transform(dtm)
```

```python
import numpy as np
sorting=np.argsort(lda.components_)[:,::-1]
features=np.array(vect.get_feature_names())
```

```python
import mglearn
mglearn.tools.print_topics(topics=range(5), feature_names=features,
sorting=sorting, topics_per_chunk=5, n_words=10)
```

```python
topic1 = np.argsort(lda_dtf[:,4])[::-1]
for i in topic1[:10]:
     print(shearss[i] + "\n")
```

```python
topic2 = np.argsort(lda_dtf[:,3])[::-1]
for i in topic2[:10]:
    print(shearss[i] + "\n")
```
